# <b> Experiment 14: Naive Bayes Classifier using Titanic dataset (We have to predict whether a person survived or not?) <b>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('titanic_data.csv')
df.head()
```

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 'columns', inplace=True)
df.head()
```

```python
target = df.Survived
input = df.drop('Survived', axis = 'columns')
```

```python
target
```

```python
input
```

```python
input.shape
```

```python
dummies = pd.get_dummies(input.Sex)
dummies.head(3)
```

```python
input = pd.concat([input,dummies],axis='columns')
input.head()
```

```python
input.shape
```

```python
input.drop('Sex', axis='columns', inplace=True)
input.head()
```

```python
input.columns[input.isna().any()]
```

```python
input.Age[:10]
```

```python
input.Age = input.Age.fillna(input.Age.median())
input.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input, target, test_size=0.2, random_state=0)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(input)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
# Accuracy : 78.21 %

# from sklearn.naive_bayes import ComplementNB 
# model = ComplementNB()
# Accuracy = 73.18 %

# from sklearn.naive_bayes import MultinomialNB
# model = MultinomialNB()
# Accuracy = 72.625 %
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```
