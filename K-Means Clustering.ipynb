# <b> Experiment 11: Implementation of K-Means Clustering Algorithm by using a Sample Dataset. <b>

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('kmeansdata.csv')
df.head(20)
```

```python
plt.scatter(df.Age, df.Income)
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Age', 'Income']])
y_predicted
```

```python
df['cluster'] = y_predicted
```

```python
df.head()
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1.Age, df1.Income, color='green')
plt.scatter(df2.Age, df2.Income, color='red')
plt.scatter(df3.Age, df3.Income, color='black')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend(['Cluster 0', 'Cluster 1', 'Cluster 2'])
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income']])
```

```python
df['Income'] = scaler.transform(df[['Income']])
df
```

```python
scaler.fit(df[['Age']])
```

```python
km = KMeans(n_clusters=3)
```

```python
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Income']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1.Age, df1.Income, color='green')
plt.scatter(df2.Age, df2.Income, color='red')
plt.scatter(df3.Age, df3.Income, color='black')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend(['Cluster 0', 'Cluster 1', 'Cluster 2'])
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1.Age, df1.Income, color='green')
plt.scatter(df2.Age, df2.Income, color='red')
plt.scatter(df3.Age, df3.Income, color='black')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='purple', marker='*', label='centroid')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend(['Cluster 0', 'Cluster 1', 'Cluster 2', 'Centroid'])
```

```python
k_rng = range(1,10)
sse = []

for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age', 'Income']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k')
plt.ylabel('Sum of Squared Error')
plt.plot(k_rng,sse)
```

### Hence, k = 3 is the best value for this Algorithm.
